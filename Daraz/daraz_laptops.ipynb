```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
```

### Scraping all daraz laptops

```python
# scraping daraz
driver=webdriver.Firefox()
price=[]
a=[]
for i in range(1,41):
    url=f"https://www.daraz.com.np/laptops/?page={i}"
    driver.get(url)
    driver.implicitly_wait(10)
    page_source=driver.page_source
    soup=BeautifulSoup(page_source,'html.parser')
    div=soup.find_all('div',class_="box--ujueT")
    for divs in div:
        a_tags=divs.find_all('a')
        spans=divs.find_all('span',class_="currency--GVKjl")
        ap=[a.text for a in a_tags]
        p=[a.text for a in spans]
        a.extend(ap)
        price.extend(p)
driver.quit()
```

```python
for pr in price:
    pr=pr.replace(",","")
```

```python
price
```

```python

laps = [item for item in a if item] # Filtering data "" esto aayera
```

```python
laptops=[]
for l ,p in zip(laps,price):
    laptops.append({"name":l,"price":p})
    
```

```python
laptops
```

### writing into csv files

```python
import csv
file=open('daraz_laptops.csv','w', encoding='utf-8')
writer=csv.DictWriter(file,fieldnames=['name','price'])
writer.writeheader()
writer.writerows(laptops)
```

```python
df=pd.read_csv('daraz_laptops.csv')
```

```python
df['price']=[int(i.replace('Rs. ',"").replace (',',""))for i in df['price']]
```

### Plotting using plotly

```python
import plotly.express as px

fig = px.density_heatmap(x=df['price'], y=df['price'], title="Laptop Price Density Heatmap")

fig.show()
```

### plotting using 2d histograms


```python
import plotly.graph_objects as go


# Creating a 2D histogram using Plotly
fig = go.Figure(data=[go.Histogram2d(x=df['price'], y=df['price'], colorscale='Viridis')])

# Adding labels and title
fig.update_layout(
    title="2D Histogram of Laptop Prices",
    xaxis_title="Price",
    yaxis_title="Price"
)

# Display the plot
fig.show()
```
